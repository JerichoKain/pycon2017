# Building Stream Processing Applications
Amit Ramesh & Qui Nguyen

### Why Stream Processing
all about data processing
- batch
- stream

Batch
- finite chunk of data
- operations defined over entire input

Stream
- unbound stream of events flowing in
- events are processed over time

why choose stream?
- Lower latency
- most data is unbound, and streaming is more flexible

### Putting an application together
example
- ad campaign metrics
 - have ads
 - count views and clicks from logs (flat json)

Components of data operations
- source
- stream processing engine
- data sink

types of operations
1) ingestion
 - data reader
2) stateless transforms
 - ie, filter, projection (ignore some fields, like map)
3) Stateful Transforms
 - care about a window of data, sliding or tumbling
 - ie, aggregation
4) keyed stateful transforms
 - group events by key (shuffle) within each window
 - ie, sum_views_by_cid, join_by_cid (multiple streams?)
5) publishing
 - writing out

### Design principles and tradeoffs
Horizontal scalability: basic idea
- data can always be broken down and distributed
- user growth or regular usage (more at day vs less at night)

How?
- operations are atomic, so partitioning
 - random partitions (multiple ingestors)
 - key partitions (sort by type)
 
watch out for hotspots, data skew

Handling failures

idempotency
- can be applied more than once and have the same effect
- ie, mutations, or stateless
- some writes can be idempotent, like unique keys.
- increments might break it.

consistency vs availability
- c = every read sees a current view of the data
- a = capacity to serve request

example: replicated data store
