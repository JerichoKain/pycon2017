 - concurrency through distrust

How to change threads

 - Virtual Memory Table

Threads and Processes

False Ideals

Windows 2001
 - simultaneous multithreading
 - 1 core, 2 threads
 - symmeteric multiprocessing
 - race conditions everywhere
  - synchronisation
  - locks, semaphores and mutexes
  - Trust but verify (Don't look before you leap)
  - test and set (atomic instructions)
 - Thread Safety
 
Windows Vista / Windows 10
 - Even more cores!
  - 4 cores, + hyperthreading
  - 12 or 16 cores + hyperthreading in servers
 - "tickless" kernels
 - Coprocessing
  - GPGPU & Friends
 
So back to python...
 - threading in python 
 - all threads share the GIL
 - so, why not just use one thread
 - Kernel-based networking
  - I/O completion ports, epoll, kqueue
 - GIL can still be pre-empted
  - can get around this with changing the affinity (increase priority, force it to be the only thread on the core)
 - Avoid the thundering herd
  - Scarce resource
  - No inter-thread communication
  - Microservers.
