New testing goal: correctness
 - make sure what we built matches specification
  - users can file taxes
  - taxes are calculated correctly
 - automated tests can't prove correctness
 - ...but humans understand, humans provide meaning
  - SME to check functionality
  - code reviews with sother developers
  
Humans 
 - Adaptable (understanding)
 - Inconsistent (bad at details.)
 - slow
 - expensive
 
Computation
 - Brittle
 - Consistent
 - fast (after it has been written)
 - cheap
  
  
Automated test prevent change
 - depends on situation (change expected once a year vs once a week)
 - preventing change has costs
  - code changes mean test changes
 - so 100% automation coverage isn't practical.
 
What happens when you run your software in production?
 - did it crash?
 - software is too complex to comprehend through testing
 - so test from the running app
  - load testing
  - collecting tracebacks and errors from production logs
  - usability testing
  - A/B testing
  - Learning about new users needs and problems
 - users are the final testers

Common mistakes
 - focusing on correct but changable
 - no learning, no feedback cycle
 - unchanging but wrong
 - preventing change but code is always changing
