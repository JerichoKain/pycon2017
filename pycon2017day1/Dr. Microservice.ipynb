# Dr. Microservice
or how I learned to stop worrying and love the API - Ryan Anguiano

Don't start a new project with microserveces
Hard to pivot from business standpoint
Core of project should be rigidly defined
Switch to microservice after you have the monolith

very large django app
old python
dependency hell (80+ req)
deployment interrupts entire app

breaking up the monolith
analyze your data flow
divide application into logical services
leave complicated business logic intact
focus on traffic (graphs)

building a migration road map
evaluate different tools (docker, mantl, kafka, ...)
use solution that best meets your needs
you can keep your existing project running along side new services

services and api design
standardize two methods of communication
sync, and async. (Http Rest and Kafka Messages)

the 12 factor app http12factor.net
declarative configuration
environment. portability
use as guide, not dogma

Services and API design
Do no make breaking changes to API endpoints
Increment endpoint version or add new endpoint
Always test every older endpoint version

Separating Data Stores
Every database should only be accessed by a single service
if a database needs to be shared, wrap it in a rest endpoint

micorservice toolset creation
auth
messaging 
log settings
Correlation IDs
middleware
...

on django, add a backent_api to request objects
user = request.backend_api.get(('accounts, 'user'))

Devops and infrastructure design
use docker to develop in the same env. as prod.
use terraform and ansible for config-based infrastructure
use CI to automate deployments

centeralized data pipeline
apache kafka
confluent open source platform
all data has a schema
producers consumers and connectors. 

logging and analytics
send all logs into data pipeline
send all docker logs into kafka using logspout
send all syslogs into kafka using kafka connect
generate correlation ides (CID) on all initial actions (aka tracing)
use CID in logs and all communications between servies
CID can be used to trace issues across entire distributed systems
